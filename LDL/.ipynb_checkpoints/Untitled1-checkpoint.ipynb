```python
name = 'Yeast_alpha'
X, ll, ld = load(name)
print("dataset name:", name)
print("# of labels:", ld.shape[1])
print("# of features:", X.shape[1])
_, subfigs = plt.subplots(6, 3, figsize=(15, 17))
for r in range(6):
    for c in range(3):
        sns.distplot(ld[:,r * 3 + c], bins=200, kde=False, ax=subfigs[r][c])
```
